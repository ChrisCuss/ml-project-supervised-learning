```python
import pandas as pd
```

```python
# import CSV file with the data
df = pd.read_csv('diabetes.csv')
```

```python
# Display first few rows
df.head()
```

```python
# Summary of dataframe
df.info()
```

```python
# Statistical summary of numerical columns
df.describe()
```

```python
# Replacing those 0s with the mean in Glucose
# Calculate the mean of 'Glucose' excluding zeros
glucose_mean = df[df['Glucose'] > 0]['Glucose'].mean()
```

```python
df.loc[df['Glucose'] == 0, 'Glucose'] = glucose_mean
```

```python
#How many people have their Glucose at 0
glucose_0 = df[df['Glucose'] == 0]
glucose_0.count()
```

```python
#How many people have their BloodPressure at 0
glucose_0 = df[df['BloodPressure'] == 0]
glucose_0.count()
```

```python
# Replacing those 0s with the mean in BloodPressure
BloodPressure_mean = df[df['BloodPressure'] > 0]['BloodPressure'].mean()
df.loc[df['BloodPressure'] == 0, 'BloodPressure'] = BloodPressure_mean
```

```python
#How many people have their SkinThickness at 0
glucose_0 = df[df['SkinThickness'] == 0]
glucose_0.count()
```

```python
# Replacing those 0s with the mean in SkinThickness
SkinThickness_mean = df[df['SkinThickness'] > 0]['SkinThickness'].mean()
df.loc[df['SkinThickness'] == 0, 'SkinThickness'] = SkinThickness_mean
```

```python
#How many people have their BMI at 0
glucose_0 = df[df['BMI'] == 0]
glucose_0.count()
```

```python
# Replacing those 0s with the mean in BMI
BMI_mean = df[df['BMI'] > 0]['BMI'].mean()
df.loc[df['BMI'] == 0, 'BMI'] = BMI_mean
```

```python
# Count of unique values in each column
df.nunique()
```

```python
# Checking the stat summaries again to make sure there aren't any more 0 values where there shouldn't bne
df.describe()
```

## Interaction Effect Between Predictor Variables

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Glucose & Age
model = ols('Outcome ~ Glucose * Age', data=df).fit()
print(model.summary())
```

```python
# BloodPressure & SkinThickness
model = ols('Outcome ~ BloodPressure * SkinThickness', data=df).fit()
print(model.summary())
```

```python
# Insuling & BMI
model = ols('Outcome ~ Insulin * BMI', data=df).fit()
print(model.summary())
```
